### Calculating Moving Averages

```python
# Calculating the 50-day and 200-day simple moving averages
df['SMA_50'] = df['close'].rolling(window = 50).mean()
df['SMA_200'] = df['close'].rolling(window = 200).mean()

#Display the first few rows with the new column
df[['close', 'SMA_50', 'SMA_200']].head()
```

### Generating Trading Signals

```python
# Generating Trading signals
df['Position'] = np.where(df['SMA_50'] > df['SMA_200'], 1, -1)
# Display the first few rows with the position column
df[['close', 'SMA_50', 'SMA_200', 'Position']].head()
```

### Visualizing the startegy 

```python
# Ploting the results
plt.figure(figsize = (14, 7))
plt.plot(df['close'], label = 'Close Price', color = 'blue')
plt.plot(df['SMA_50'], label = '50-day SMA', color = 'orange')
plt.plot(df['SMA_200'], label = '200-day SMA', color = 'green')
plt.fill_between(df.index, df['close'], where = (df['Position'] > 0), color = 'lightgreen', alpha = 0.3, label = 'Long Position')
plt.fill_between(df.index, df['close'], where = (df['Position'] < 0), color = 'lightcoral', alpha = 0.3, label = 'Long Position')
plt.title('Title Following Strategy: Moving Average Crossover')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

# 2. Mean Reversion

```python
# Define the ticker list
tickers_list = ['AAPL', 'MSFT']

# Set up the API key and headers
headers = {
    'Content-Type': 'application/json'
}

# Define functions to fetch data from Tiingo
def fetch_tiingo_data(ticker, start_date = '2023-01-01'):
    url = f'http://api.tiingo.com/tiingo/daily/{ticker}/prices?startDate={start_date}&token={api_key}'
    response = requests.get(url, headers = headers)
    data = response.json()

    # Convert the response into a DataFrame
    df = pd.DataFrame(data)
    df['date'] = pd.to_datetime(df['date'])
    df.set_index('date', inplace = True)
    return df['adjClose']

# Fetch the data for both tickers
data = pd.concat([fetch_tiingo_data(ticker) for ticker in tickers_list], axis = 1)
data.columns = tickers_list

# print the first 5 rows of the data
print(data.head())
```

### Calculating the Spread

```python
import numpy as np
import matplotlib.pyplot as plt

# Calculate the spread
spread = data['AAPL'] - data['MSFT']

# Normalize spread by substracting mean and dividing by standard deviation
spread_normalized = (spread - spread.mean()) / spread.std()

# Plot the normalized spread
plt.figure(figsize = (10, 6))
plt.plot(spread_normalized, label = 'Normalized Spread')
plt.title('Normalized Spread between AAPL and MSFT')
plt.xlabel('Date')
plt.ylabel('Normalized Spread')
plt.legend()
plt.show()
```

### Generating Trading Signal

```python
# Creating trading signals
shorts = spread_normalized > 1
longs = spread_normalized < -1

# print signals
print(shorts, longs)
```

```python
# create exit signals
exits = spread_normalized * spread_normalized.shift(1) < 0
# Print exit signals
print(exits)
```

## Backtest the strategy

```python
# Calculate daily returns
returns = pd.Series(index = data.index, dtype = 'float64')
returns[longs] = data['AAPL'].shift(-1)[longs] / data['AAPL'][longs] - 1
returns[shorts] = -(data['AAPL'].shift(-1)[shorts] / data['AAPL'][shorts] - 1 )
returns[exits] = 0

# Calculate cumulative returns
cumulative_returns = (1 + returns).cumprod()

# Plot the cumulative returns
cumulative_returns.plot()
```
