### Aggregating Monthly Returns

```python
def aggregated_returns(returns, time_span):
    def calculated_cumulative_returns(x):
        return (1+x).cumprod()[-1] - 1.0
    if time_span == 'yearly':
        return returns.groupby(lambda x: x.year).apply(calculated_cumulative_returns)
    elif time_span == 'monthly':
        return returns.groupby([lambda x: x.year, lambda x: x.month]).apply(calculated_cumulative_returns)
```

```python
mr = aggregated_returns(data['Returns'], 'monthly')
```

```python
mr.head()
```

```python
mr_transpose = mr.unstack()
mr_transpose 
```

```python
import numpy as np
final_mr = np.round(mr_transpose, 3)
final_mr
```

```python
final_mr.rename(columns = {
         1: 'Jan', 2: 'Feb', 3: 'Mar',
         4: 'Apr', 5: 'May', 6: 'Jun',
         7: 'Jul', 8: 'Aug', 9: 'Sep',
         10: 'Oct', 11: 'Nov', 12: 'Dec'
}, inplace = True)
final_mr
```

```python
from matplotlib import cm
import seaborn as sns
sns.heatmap(final_mr * 100, 
            annot = True, #Label each block in our heatmap with the value of our monthly returns.
            annot_kws = {'size': 8}, #Change the font size of our annotations.
            alpha = 1.0, #Control the translucency of the colour in each block of the heatmap. In this case there is no transparency.
            center = 0, #Control the colour of data blocks that correspond to a value of 0. Here we make the center yellow to represent months where we have 0.0% returns.
            cmap = cm.RdYlBu
           )
plt.show()
```
